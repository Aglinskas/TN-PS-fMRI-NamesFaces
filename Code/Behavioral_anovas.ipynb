```python
fn = '/Users/aidasaglinskas/Desktop/TN-PS-fMRI-NamesFaces/Data/behav_data/RT_table_full.csv'
data = pd.read_csv(fn)
data
```

```python
pg.rm_anova(data,correction=True,detailed=True)
```

```python
grouped_arr = np.hstack((
    data[['Attractiveness','Distinctiveness']].values.mean(axis=1)[:,np.newaxis],
    data[['Friendliness','Trustworthiness']].values.mean(axis=1)[:,np.newaxis],
    data[['First memory','Familiarity']].values.mean(axis=1)[:,np.newaxis],
    data[['How many facts','Occupation']].values.mean(axis=1)[:,np.newaxis],
    data[['Common Name','Common Surname']].values.mean(axis=1)[:,np.newaxis]
))
```

```python
grouped_data = pd.DataFrame(grouped_arr,columns=['Physical','Social','Episodic','Biographical','Nominal'])
grouped_data
```

```python
pg.rm_anova(grouped_data,correction=True,detailed=True)
```
